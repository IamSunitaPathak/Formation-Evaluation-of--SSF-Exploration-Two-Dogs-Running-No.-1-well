# VShale

```python
GRmin = 29
GRmax = 120.2

df['VShale'] = ((df.GR - 29.0 ) / (120.2 - 29.0))
df['PHIX_Sand']  = (df.PHID_SAND + df.NSAND)/2
df['VSand'] = 1 - (df.VShale) - (df.PHIX_Sand)
df.round(decimals = 2)
```

```python
df['PHI_Eff'] = df.PHIX_Sand * (1 - df.VShale)

df['VSand_correct'] = 1 - df.PHI_Eff - df.VShale
df.round(decimals = 2)
df.head(70).round(decimals = 2)
```

```python
fig, ax = plt.subplots(figsize = (20,5))
ax.stackplot(df.Depth_m, df.VShale, df.VSand_correct, df.PHI_Eff, colors = ('black', 'lemonchiffon', 'azure'), labels = ('Shale', 'Sand', 'Porosity'))

plt.xlabel('Depth', size = 20)
plt.title('Porosity Corrected Lithology Plot', size = 25, color = 'red')

plt.legend(loc='upper center', bbox_to_anchor=(1.07, 1.05), fancybox=True, shadow=True, fontsize = 15)
plt.show()
```

```python
a = 0.62
m = 2.15
n = 2

df['Rwa'] = df.PHI_Eff**m * df.ILD_Ohm_m

Rwa = 0.07                       # Looking at the curve below

df['Sw'] = ((a / df.PHI_Eff**m) * ( Rwa / df.ILD_Ohm_m))**0.5
df['Sw'].values[df['Sw'].values > 1.0 ] = 1.0
df['Shc'] = 1 - df.Sw
df
```

```python
plt.figure(figsize = (18,8))

plt.subplot(1,4,1)
plt.plot(df.PHI_Eff, df.Depth_m)

plt.xlabel('PhiEff', size = 20)
plt.ylabel('Depth', size = 20)
plt.title('Depth vs PhiEff', size = 20, color = 'red')

plt.xlim(0.5,0.0)
plt.ylim(2620, 2545)

#********************** Rwa ************************8

plt.subplot(1,4,2)
plt.plot(df.Rwa, df.Depth_m, color = 'red')

plt.xlabel('Rwa', size = 20)
plt.ylabel('Depth', size = 20)
plt.title('Depth vs Rwa', size = 20, color = 'green')

plt.xlim(1,0.01)
plt.ylim(2620, 2545)

plt.xscale('log')

#********************** Sw **********************

plt.subplot(1,4,3)
plt.plot(df.Sw, df.Depth_m, color = 'blue')

plt.xlabel('Sw', size = 20)
plt.ylabel('Depth', size = 20)
plt.title('Depth vs Sw', size = 20, color = 'blue')

plt.xlim(1,0.01)
plt.ylim(2620, 2545)

#********************** Shc **********************

plt.subplot(1,4,4)
plt.plot(df.Shc, df.Depth_m, color = 'black')

plt.xlabel('Shc', size = 20)
plt.ylabel('Depth', size = 20)
plt.title('Depth vs Shc', size = 20, color = 'black')

plt.xlim(1,0.01)
plt.ylim(2620, 2545)

plt.subplots_adjust(wspace = 1)


plt.show()
```

```python
df['Rmfa'] = df['PHI_Eff']**m * df['MSFL'] / a
df['Rmfa']
```

```python
Tgrad = 2.678 # C/100m
Tms = 15

df['Tform_C'] = Tms + (Tgrad * df.Depth_m / 100)
df['Tform_F'] = df.Tform_C * 9/5 + 32
df['K'] = (df['Tform_F'] + 505)/8

Rw = 0.06

df['Rmf_SP'] = Rw * (10**(-df['SP_mV'] / df['K'] ))
df.columns
```

```python
plt.figure(figsize = (5,10))

plt.plot(df['Rmfa'], df['Depth_m'], color = 'orange', label = 'Rmfa')
plt.plot(df.Rmf_SP, df.Depth_m, color = 'blue', label = 'Rmf_SP')

plt.xlim(0.01, 1)
plt.ylim(2620, 2545)

plt.rcParams['ytick.right'] = plt.rcParams['ytick.labelright'] = False
plt.rcParams['ytick.left'] = plt.rcParams['ytick.labelleft'] = True

plt.xscale('log')
plt.legend()

plt.title('Rmf-Archie & SP', color = 'red', size = 20)
plt.show()
```

```python
a = 0.62
m = 2.15
n = 2
Rmfa = 0.25

df['Sxo'] = ((a / (df['PHI_Eff'])**m) * (Rmfa / df['MSFL']))**0.5

df['Sxo_correct'] = df['Sxo']
df['Sxo_correct'].values[df['Sxo_correct'].values > 1.0] = 1.0
```

```python
plt.figure(figsize = (5,10))

plt.plot(df['Sxo_correct'], df['Depth_m'], color = 'orange', label = 'Sxo_correct')
plt.plot(df['Sw'], df['Depth_m'], color ='blue', label = 'Sw')

plt.ylim(2620, 2540)

plt.title('Sw vs Swo', color = 'red', size = 25)
plt.legend(loc = 'upper center')

plt.show()
```

```python
df['BVW'] = df['Sw'] * df['PHI_Eff'] 
df['BVHC'] = df['Shc'] * df['PHI_Eff'] 
```

```python
df['Gas'] = df['Shc']
df['Gas'].values[(df['PHID_SAND'] - df['NSAND']).values < 0.03] = 0

df['Oil'] = df['Shc'] - df['Gas']
```

```python
plt.figure(figsize = (5,10))

plt.plot(df['Gas'], df['Depth_m'], color = 'red', label = 'Gas' )
plt.plot(df['Oil'], df['Depth_m'], color = 'green', label = 'Oil')

plt.xlim(-0.2,1,0.1)
plt.ylim(2620, 2540)

plt.title('Gas vs Oil Saturations', color = 'red', size = 20)
plt.legend()

plt.show()
```

```python
df.columns
```

```python
df['BVG'] = df['PHI_Eff'] * df['Gas']
df['BVO'] = df['PHI_Eff'] * df['Oil']
```

```python
fig, ax = plt.subplots(figsize = (20,5))
ax.stackplot(df.Depth_m, df.VShale, df.VSand_correct, df['BVW'], df['BVG'], df['BVO'], colors = ('black', 'yellow','blue', 'red', 'lightgreen'), labels = ('Shale', 'Sand', 'Water', 'Gas', 'Oil'))

plt.xlabel('Depth', size = 20)
plt.title('Lithology Plot Of the Core Sample', size = 25, color = 'red')

plt.legend(loc='upper center', bbox_to_anchor=(1.07, 1.05), fancybox=True, shadow=True, fontsize = 15)

plt.show()
```
